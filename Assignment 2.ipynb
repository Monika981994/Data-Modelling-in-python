```python
#change the categorical variable salary to numerical
sal = {"low": 0, "medium": 1, "high": 2}
turnover['sal_int'] = turnover['Salary'].map(sal)
```

```python
#change the categorical variable sales to numerical
sale={"sales":0, "accounting":1, "hr":2, "technical":3, "support":4, "management":5,
        "it":6, "product_mng":7, "marketing":8, "randd":9}
turnover['sale_int'] = turnover['Department'].map(sale)
```

```python
#delete the field salary and sales from original dataset
turnover.drop('Salary', inplace=True, axis=1)
turnover.drop('Department', inplace=True, axis=1)
```

```python
#Split features and target
turnover_f=turnover[['Satisfaction_Level','Last_Evaluation','Number_Of_Projects','Avg_Month_Hrs','Time_Spend_Company'
                             ,'Work_Accident','Promotions','sal_int','sale_int']]
turnover_t=turnover['Left']
```

```python
from sklearn.feature_selection import RFE
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(5,metric='manhattan',p=1,leaf_size=40,weights='distance',algorithm='kd_tree')

rfe = RFE(model, 10)
rfe = rfe.fit(turnover_f,turnover_t)
print(rfe.support_)
print(rfe.ranking_)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(turnover_f,turnover_t,test_size=0.30,random_state=6)
```
